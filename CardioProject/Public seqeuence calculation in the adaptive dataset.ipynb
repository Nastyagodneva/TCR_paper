## define a function that calculates the percent of shared sequences in a dataset, given the sample names, number of samples to select, number of sequences to select from each sample, and number of calculation repeats. 

(1) choose randomly nSamples
(2) for each sample, load its uniqueAA df, and randomly choose nSequences, and add them to the new uniqueAA df
(3) count values for the new uniqueAA sequences
(4) calculate percent of public and save value
(5) repeat for nRepeats and calculate the mean.



```python
#this function calculates the percent of shared sequences in a dataset, given the 

def testPublicRatesInSelectedSamples(sample_filenames,nSamples,nSequences,nRepeats,useSequenceWeights):
    
    percPublicDF=pd.DataFrame()
    for r in range(nRepeats):
        print 'starting repeat %s...' %r
        
        # (1) choose randomly 'nSamples' samples (currently nSamples=434)
        print 'choosing %s samples...' %nSamples
        selectedSamples=np.random.choice(sample_filenames, nSamples, replace=False)

        #(2) for each sample, load its uniqueAA df, and randomly choose 'nSequences' sequences, and add them to the new uniqueAA df
        # (currently nSequences=12,756)
        print 'loading uniqueAA df selecting %s sequences and adding to the large uniqueAA df...' %nSequences
        TotaluniqueAA=pd.DataFrame()
        for n, sample_name in enumerate(selectedSamples):
                print n, sample_name
                file1='/net/mraid08/export/genie/Lab/Personal/ShaniBAF/%s/uniqueAAdfs/%s' %(data_folder,sample_name)
                uniqueAA=pd.read_pickle(file1)
                uniqueAAlist=list(uniqueAA.index)
                if useSequenceWeights:
                    countsList=list(uniqueAA[('count (templates)','sum')])
                    countsListSum=np.sum(countsList)
                    countsListNorm=[float(x)/countsListSum for x in countsList]
                else:
                    countsListNorm=None
                if len(uniqueAAlist)>nSequences:
                    chosenUniqueAAs= np.random.choice(uniqueAAlist, nSequences, replace=False,p=countsListNorm)
                else:
                    chosenUniqueAAs= np.random.choice(uniqueAAlist, len(uniqueAAlist), replace=False,p=countsListNorm)
                    print '% seqeunces only were taken from sample %s' %(len(uniqueAAlist),sample_name)
                chosenUniqueAAsDF=pd.DataFrame({'aminoAcid':chosenUniqueAAs,'Sample': sample_name})
                TotaluniqueAA=pd.concat([TotaluniqueAA,chosenUniqueAAsDF])
        
        #(3) count values for the new uniqueAA sequences:
        print 'counting values for uniqueAA sequences in repeat %s' %r
        TotaluniqueAA=TotaluniqueAA.set_index('aminoAcid')
        CountSharing=TotaluniqueAA.index.value_counts()
        
        
        #(4) calculate percent of public and save value:
        print 'calculating percent of public sequences and saving value for repeat %s' %r
        CountSharing=pd.DataFrame(CountSharing)
        CountSharing=CountSharing.rename(columns={'aminoAcid':'n_shared_samples'})
        Npublic=len(CountSharing[CountSharing['n_shared_samples']>1])
        Nprivate=len(CountSharing[CountSharing['n_shared_samples']==1])
        percPublic=float(Npublic)*100/Nprivate
        
        print 'percPublic in repeat %s is %s' %(r,percPublic)
        percPublicDF.loc[r,'percPublic']=percPublic
        
    return percPublicDF
    
        
        
    
```

## calculate percPublic for 434 samples, 12576 sequences per sample, with 10 calculation repeats:

```python
sample_filenames=sample_filenames
nSamples=434
nSequences=12576
nRepeats=10

percPublicDF=testPublicRatesInSelectedSamples(sample_filenames,nSamples,nSequences,nRepeats)
file1='/net/mraid08/export/genie/Lab/Personal/ShaniBAF/TCR_real_data/PublicSeqAnalysis/percPublicCalc_AdaptiveDataSimulation_%s_%s_%s' %(nSamples,nSequences,
                                                                                                                                  nRepeats)
percPublicDF.to_pickle(file1)
```
