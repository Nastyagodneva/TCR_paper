```python
def edit_summaryDF(summary_df,DirToSave=False,count='',nonbinary_name='',binary_name='',sample_list_name=''):
    summary_df=add_corrected_pValues(summary_df,pValueColumn='MW_p',nTests=summary_df.shape[0],FDR=0.1)
    
    summary_df=add_corrected_pValues(summary_df,pValueColumn='fisher_p',nTests=summary_df.shape[0],FDR=0.1)
    summary_df=summary_df.sort_values(by='MW_p')
    
    if DirToSave is not None:
        if not isdir(DirToSave):
            makedirs(DirToSave)
        summary_df.to_excel(DirToSave + '%s_%s_%s_associations_%s.xlsx' %(nonbinary_name,binary_name,count,sample_list_name))
    
    return summary_df

def analyze_associations_binary_nonbinary(nonbinaryDF,binaryDF,nShared_pernonbinary,nShared_perbinary,
                                          nonbinary_zero=0.0001,nonbinary_name='MbSpecies',binary_name='Seq',make_concise=True):
    
    #this function is a generalized form of the function analyze_tcr_Mb_relations
    # in this function, the nonbinary can be microbiome relative abundance table or any other nonbinaryotypic table
    # the function will calculate the fisher and mann-whitney associations between each binary sequence/cluster in the binaryDF table, which is binary,
    # and the nonbinaryDF table which is not binary
    # for fisher test, a binary cut off for the nonbinary should be selected. for continuous nonbinarys, consdier using the mean or median. for microbiome 
    #RA data, this is usually 0.0001
    ########
    
    #filter nonbinary and generate binary df:
    sample_count=nonbinaryDF.count()
    nonbinary_filtered=nonbinaryDF.loc[:,sample_count[(sample_count>=nShared_pernonbinary)&(sample_count<len(nonbinaryDF)+1)].index].fillna(0)
    print ('nonbinaryDF shape after filtering only for nonbinary that appear in more than %s: ' %nShared_pernonbinary,nonbinary_filtered.shape)
#     print nonbinary_filtered.iloc[:4,:4]
    nonbinary_filtered_binary=(nonbinary_filtered>nonbinary_zero).astype(int)
    print ('binary:')
#     print nonbinary_filtered_binary.iloc[:4,:4]
    
    #filter binarys:
    cols_to_use=[col for col in binaryDF if (binaryDF[col].sum()>=nShared_perbinary) & (binaryDF[col].sum()<len(binaryDF))]
    binaryDF_filtered=binaryDF[cols_to_use]
    print ('binaryDF shape after filtering only for seqs that appear in more than %s: ' %nShared_perbinary,binaryDF_filtered.shape)
#     print binaryDF_filtered.iloc[:4,:4]

    print ('total number of tests: ',binaryDF_filtered.shape[1]*nonbinary_filtered.shape[1] )
    
    #run over all seqs and nonbinary and calculate MW and fisher test:
    summary_df=pd.DataFrame()
    count=0
    for i, seq in enumerate(binaryDF_filtered.columns):
        print (i, seq)
    #     print i,seq
        for j, s in enumerate(nonbinary_filtered.columns):
            if j%1000 == 0: print (str(j) + '  ' +str(datetime.datetime.now()))
            
            #calculate mann-whitney:
            merged1=pd.merge(pd.DataFrame(binaryDF_filtered[seq]),pd.DataFrame(nonbinary_filtered[s]),
                                            how='inner',left_index=True,right_index=True)
#             print ('merged1 shape: ', merged1.shape)
#             print 'merged1'
#             print merged1.head()
            data={}
            names=[]
            for name,group in merged1.groupby(seq):
                data[name]=group[s].tolist()
                names.append(name)
            try:
                MW_s, MW_p=mannwhitneyu(data[names[0]],data[names[1]])          
                mean_nonbinaryFreq_seqAbs=np.mean(data[0])
                mean_nonbinaryFreq_seqPres=np.mean(data[1])
            except:
                print ('MW test failed, probably this clusters doesnt exist in samples')
                MW_s=np.nan; MW_p=np.nan; mean_nonbinaryFreq_seqAbs=np.nan; mean_nonbinaryFreq_seqPres=np.nan
            
            #calculate Fisher:
            merged2=pd.merge(pd.DataFrame(binaryDF_filtered[seq]),pd.DataFrame(nonbinary_filtered_binary[s]),
                                            how='inner',left_index=True,right_index=True)
#             print 'merged2'
#             print merged2.head()
            if (merged2[s].sum()<len(merged2)) | (merged2[seq].sum()<len(merged2)):
                tab = pd.crosstab(merged2[seq],merged2[s], dropna=False).fillna(0)
                binary0_nonbinary0=tab.iloc[0,0]
                binary0_nonbinary_1=tab.iloc[0,1]
                binary1_nonbinary0=tab.iloc[1,0]
                binary1_nonbinary1=tab.iloc[1,1]
                
    #             print (tab)
                try:
                    fisher_OR, fisher_p = fisher_exact(tab, alternative='two-sided')
                except:
                    print ('couldnt execute fisher test for seq %s and nonbinary %s' %(seq,s))
#                     print (tab)
                    fisher_OR= 999999; fisher_p=1
            else:
                fisher_OR= 999999; fisher_p=1
                binary0_nonbinary0=np.nan; binary0_nonbinary_1=np.nan
                binary1_nonbinary0=np.nan; binary1_nonbinary1=np.nan
              
            #update summary_df:
            summary_df.loc[count,binary_name]=seq
            summary_df.loc[count,nonbinary_name]=s
            summary_df.loc[count,'mean_%sFreq_%s_0' %(nonbinary_name,binary_name)]=mean_nonbinaryFreq_seqAbs
            summary_df.loc[count,'mean_%sFreq_%s_1' %(nonbinary_name,binary_name)]=mean_nonbinaryFreq_seqPres
            summary_df.loc[count,'MW_s']=MW_s
            summary_df.loc[count,'MW_p']=MW_p
            summary_df.loc[count,'MW_s']=MW_s
            summary_df.loc[count,'MW_p']=MW_p
            summary_df.loc[count,'%s_0_%s_0' %(binary_name,nonbinary_name)]=binary0_nonbinary0
            summary_df.loc[count,'%s_0_%s_1' %(binary_name,nonbinary_name)]=binary0_nonbinary_1
            summary_df.loc[count,'%s_1_%s_0' %(binary_name,nonbinary_name)]=binary1_nonbinary0
            summary_df.loc[count,'%s_1_%s_1' %(binary_name,nonbinary_name)]=binary1_nonbinary1
            summary_df.loc[count,'fisher_OR']=fisher_OR
            summary_df.loc[count,'fisher_p']=fisher_p

            #update count:
            count=count+1
            
            if count%20000 == 0:
                print ('editing and saving temp summary_df...')
                           
                summary_df = edit_summaryDF(summary_df,DirToSave=TCR_MB_DIR+'temp/',count=count,nonbinary_name=nonbinary_name,binary_name=binary_name,sample_list_name=len(binaryDF))

    summary_df = edit_summaryDF(summary_df,DirToSave=TCR_MB_DIR+'temp/',count=count,nonbinary_name=nonbinary_name,binary_name=binary_name,sample_list_name=len(binaryDF))
    
    if make_concise:
        concise_summary_df=summary_df[[binary_name,nonbinary_name,'fisher_OR','MW_p','MW_p_corrPval_FDR0.1']]
        concise_summary_df['-log10p_MW'] = -np.log10(concise_summary_df['MW_p'])
        concise_summary_df['Associated_with_%s_positive' %binary_name] = np.where(concise_summary_df['fisher_OR']>=1,1,0)
        concise_summary_df=concise_summary_df.drop(['fisher_OR','MW_p'],axis=1)
    else: concise_summary_df=None


    return summary_df,concise_summary_df
```

## get data
only balanced samples, which have swab Mb data, 
only clusters that appear in at least 5% of samples in at least one cohort

```python
def get_cluster_data(min_shared=0.05,max_shared=1,PNP530_balanced_samples_with_swab_data=None, Cardio126_balanced_samples_with_swab_data=None):
        
    #get cluster data - all samples, 0.5-0.9 percent
    data_file='sampleByClusterDF_cohortfiltering005-09perc_dropped.dat'
    data = pd.read_pickle('/net/mraid08/export/jafar/Microbiome/Analyses/ShaniBAF/sharingAnalysis/seqClusters_allProd_maxdist1/' + data_file)
    print ('data.shape: ', data.shape)
#     print ('data.head: ', data.iloc[:4,:4])
           
    #get data only for clusters that appear in at least 5%  of the samples in each chort (balanced cohorts:):
    if PNP530_balanced_samples_with_swab_data is not None:  PNP530_balanced = PNP530_balanced_samples_with_swab_data
    if Cardio126_balanced_samples_with_swab_data is not None:  Cardio126_balanced = Cardio126_balanced_samples_with_swab_data
        
    cohort_list = [['PNP530',PNP530_balanced],['Cardio126',Cardio126_balanced]]
    for cohort in cohort_list:
        cohort_name=cohort[0]; sample_list=cohort[1]
        print ('')
        print ('cohort_name: ', cohort_name)
           
        data_balanced=data[data.index.isin(sample_list)]
        print ('balanced_data.shape= ',data_balanced.shape)
        
        cols = get__limited_sharing(data_balanced,min_shared,max_shared).columns.tolist()
        
#         cols=[col for col in data_balanced.columns if (data_balanced[data_balanced[col]>0.0001]).shape[0] > 0.05*]
        data_balanced_5more = data_balanced.loc[:,cols]
        print ('data_balanced_5more.shape: ',data_balanced_5more.shape)
                                                   
        cohort.append(data_balanced_5more)
        
    PNP530_data_balanced_5more = cohort_list[0][2]
    Cardio126_data_balanced_5more = cohort_list[1][2]
    
    
    # get final data df  -PNP530 AND Cardio126 together, only balanced samples, only clusters that appear in at least 5 samples in
    # at least one cohort, and only clusters that do not appear in more than 85% of total samples
    
    final_data= data.loc[PNP530_balanced+Cardio126_balanced,list(set(PNP530_data_balanced_5more.columns.tolist()+Cardio126_data_balanced_5more.columns.tolist()))]
    print ('final_data.shape: ',final_data.shape)
    final_data = get__limited_sharing(final_data,0,0.85)
    print ('final_data after removing clusters that appear in more than 85% samples.shape: ',final_data.shape)
                                                       
    return PNP530_data_balanced_5more,Cardio126_data_balanced_5more,final_data
           
    
```

# get all species, genus, clusters associated with the disease or the healthy state with FDR<0.2 (?)

```python
def get_sig_results_Mb_or_clusters(concise_summary_df,original_data_df,corr_p_treshold=None,topN=None,nonbinary_name='',sample_list=None):
    
    #get list of significant species/genus/clusters:
    
    if topN is not None:
        sig_results = (concise_summary_df[:topN][nonbinary_name]).tolist()
    else:
        sig_results=concise_summary_df[concise_summary_df['MW_p_corrPval_FDR0.1'] < corr_p_treshold][nonbinary_name].tolist()
    print ('number of significant results with FDR threshold=%s' %corr_p_treshold,len(sig_results))
    
    
    #get original data for the significant entities (can be for balanced_withSwab, all samples ot anything else):
    sig_df= original_data_df.loc[:,sig_results]
    
    if sample_list is not None: sig_df =sig_df.loc[sample_list,:]
    
    
    return sig_df
```

```python
def make_custome_legend(color_list,marker_list,label_list,ax,loc=None,bbox_to_anchor=None,**kwargs):
    
    '''
    color list = list of colors for markers
    marker_list = list of marker shapes such as 'o','*' etc.
    label_list = list of labels (strings)
    loc = legend location (string) such as 'upper right'
    bbox_to_anchor = 2-tuple indicating x,y coords for legend. see 
    https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.legend.html
    for more explanations
    **kwargs  - such as fontsize, title, title_fontsize...
    
    '''
    import matplotlib.pyplot as plt
    from matplotlib.legend_handler import HandlerBase
    
    class MarkerHandler(HandlerBase):
        def create_artists(self, legend, tup,xdescent, ydescent,
                        width, height, fontsize,trans):
            return [plt.Line2D([width/2], [height/2.],ls="",
                       marker=tup[1],color=tup[0], transform=trans)]


    ax.legend(list(zip(color_list,marker_list)), label_list, 
          handler_map={tuple:MarkerHandler()},loc=loc,bbox_to_anchor=bbox_to_anchor, **kwargs) 
    
    return ax
```

```python
def plot_cluster_Mb_disease_relations2(clusterDF,mbDF,cluster_name,mb_name,ax=None,use_binary_cluster=True,cluster_as_cat=True):
    
    
    '''
    new function that uses seaborn's stripplot.
    clusterDF is a sample by cluster dataframe (not binary)
    mbDF is a sample by species dataframe (not binary) p robably can use also genus here.
    cluster_name  = cluster head sequence (string)
    mb_name = formal species name (string)
    use_binary_cluster - whether to binarize the cluster information (to become False/True whether there is at least
    one sequence of the cluster in this sample). use false only if cluser_as_cat is True. 
    cluster_as_cat - bool, whether to use the cluster info as the x-axis variable (False means to use the isCardio info instead)
    
    '''
    merged=pd.merge(pd.DataFrame(clusterDF[cluster_name]),pd.DataFrame(mbDF[mb_name]),how='inner',left_index=True,right_index=True)
    y_values=merged[cluster_name].unique().tolist()
    merged['isCardio'] = np.where(merged.index.str.replace('BD','').astype(int) < 950,'Healthy','Patients')
    
    #fedine cluster info column to use: original or binary version
    if use_binary_cluster:
        merged['%s_binary' %cluster_name] = merged[cluster_name]>0
        cluster_col='%s_binary' %cluster_name
        xlabel_cluster = 'TCRseq cluster present'
    else:
        cluster_col=cluster_name
        xlabel_cluster = '# TCRseq in cluster'
        
    # define which column will be the x axis column:  
    if cluster_as_cat:
        x_col = cluster_col
        split_col = 'isCardio'
        xlabel = xlabel_cluster
        legend_title = ''
        color_list= ['grey','darkred']
    else:
        x_col = 'isCardio'
        merged['%s_binary' %cluster_name] = merged['%s_binary' %cluster_name].astype('str')
        split_col=cluster_col
        xlabel = ''
        legend_title = 'TCRseq cluster present'
        color_list=['forestgreen','red']
        
    if ax is None:
        fig,ax=plt.subplots()
        
      
    
    cmap=mpl.colors.ListedColormap(color_list, name='from_list', N=2)
    mpl.cm.register_cmap('from_list', cmap)
    cpal = sns.color_palette('from_list', n_colors=2)

        
    ax = sns.stripplot(x=x_col, y=mb_name, hue=split_col, data=merged,split=True,ax=ax,palette=cpal,size=10,alpha=0.7)
#     ax = sns.violinplot(x=x_col, y=mb_name, hue=split_col, data=merged,ax=ax,color='red')
    
    ax.set_ylabel('Species Relative Abundance')
    ax.set_xlabel(xlabel)
    
    marker_list=['o','o']
    label_list=list(merged[split_col].unique())
    bbox_to_anchor = (1.01,0.99)
    loc = 'upper left'
    make_custome_legend(color_list,marker_list,label_list,ax,title=legend_title,loc=loc,bbox_to_anchor=bbox_to_anchor)
    
    mb_name_species = mb_name.split('|')[-4]
    mb_name_genus = mb_name.split('|')[-5]

    ax.set_title('%s\n%s\n%s' %(cluster_name,mb_name_species,mb_name_genus),y=1.03)
    ax.set_yscale('log')
    ax.set_ylim(0.00008,0.1)
    return ax
    
    
```

```python
def plot_cluster_Mb_disease_relations(clusterDF,mbDF,cluster_name,mb_name,ax=None):
    
    merged=pd.merge(pd.DataFrame(clusterDF[cluster_name]),pd.DataFrame(mbDF[mb_name]),how='inner',left_index=True,right_index=True)
    y_values=merged[cluster_name].unique().tolist()
    merged['cluster_up']=merged[cluster_name]+0.1
    merged['cluster_down']=merged[cluster_name]-0.1
    merged_healthy=merged[merged.index.str.replace('BD','').astype(int) < 950]
#     print (merged_healthy.iloc[:4,:])
    merged_patients=merged[merged.index.str.replace('BD','').astype(int) > 949]
    
    if ax is None:
        fig,ax=plt.subplots()
    merged_healthy.plot.scatter(x=mb_name,y='cluster_up',color='grey',ax=ax,s=100,alpha=0.4)
    merged_patients.plot.scatter(x=mb_name,y='cluster_down',color='darkred',ax=ax,s=100,alpha=0.4)
#     ax.set_yticks(y_values)
    
    return ax
    
    
```
