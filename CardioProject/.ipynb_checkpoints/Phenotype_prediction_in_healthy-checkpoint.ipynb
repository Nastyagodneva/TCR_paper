```python
def calc_pred_performance(pred_and_real_df,task,pred_col,phen,binary_threshold=0.5):
    
    r_pear,p_pears, r_spear,p_spears,prevalence, AUC,precision_Recall,kappa,f1score,f1scoreMacro = np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan
    
    print ('pred col, phen: ',pred_col,phen )
    if task=='reg':


        #get spearman and pearson correlations:
        r_pear,p_pears=MyPearsonr(pred_and_real_df[pred_col],pred_and_real_df[phen])
        print ('r_pear,p_pears: ',r_pear,p_pears)
        r_spear,p_spears=MySpearmanr(pred_and_real_df[pred_col],pred_and_real_df[phen])
        print ('r_spear,p_spears: ',r_spear,p_spears)
        
    else:
         #### calculate AUC and PR:
        prevalence = float(pred_and_real_df[phen].sum()) / pred_and_real_df[phen].shape[0]
        # AUC
        fpr, tpr, thresholds = metrics.roc_curve(pred_and_real_df[phen], pred_and_real_df[pred_col], pos_label=1)
        AUC = metrics.auc(fpr, tpr)
        precision, recall, _ = precision_recall_curve(pred_and_real_df[phen], pred_and_real_df[pred_col])
        precision_Recall = metrics.auc(recall, precision)
        
        print ('Binary probability threshold=%s' %binary_threshold)
        pred_and_real_df[pred_col+'_binary']=[0 if x<binary_threshold else 1 for x in list(pred_and_real_df[pred_col])]    
        # Cohen's Kappa:
        kappa=metrics.cohen_kappa_score(pred_and_real_df[phen],pred_and_real_df[pred_col+'_binary'])
        #f1 scores:
        f1score=metrics.f1_score(pred_and_real_df[phen],pred_and_real_df[pred_col+'_binary'])
        f1scoreMacro=metrics.f1_score(pred_and_real_df[phen],pred_and_real_df[pred_col+'_binary'],average='macro')
        
        print ('prevalence: ',prevalence)
        print ('ROC_auc: ',AUC)
        print ('precision_Recall: ',precision_Recall)
        print ("cohen's kappa: ",kappa)
        print ('f1score:', f1score)
        print ('f1scoremacro:' ,f1scoreMacro)
        
    return r_pear,p_pears, r_spear,p_spears,prevalence,AUC,precision_Recall,kappa,f1score,f1scoreMacro

```

```python
def predict_phen_for_patients_basedOnHealthy_1split(phen,resultFolder,X_file,Y_file,split,task,binary_threshold=0.5,useSelectedFeatures=False,useScaler=False):
    
    
    
    ### prediction:
    
    #get saved model:
    phenotype_split_model_file='/net/mraid08/export/jafar/Microbiome/Analyses/ShaniBAF/\
predictions2/PNP530_majorPhenotypes/%s/%s_split%s_model.pkl' %(resultFolder,phen,split)
    phenotype_split_model=pickle.load(open(phenotype_split_model_file,"rb"))
       

    #get feature matrix for patients only:
    try:
        X=pd.read_pickle(X_file)
    except:
        X=pd.read_excel(X_file).set_index('BD')

#     print X.head()
    X_patients=X.loc[Cardio126,:]
    
    if useSelectedFeatures:
     #get saved selected_features and apply:
        selected_features_split_file='/net/mraid08/export/jafar/Microbiome/Analyses/ShaniBAF/\
predictions2/PNP530_majorPhenotypes/%s/%s_split%s_selectedFeatures.pkl' %(resultFolder,phen,split)
        phenotype_split_selectedFeatures=pickle.load(open(selected_features_split_file,"rb"))
        X_patients=X_patients[phenotype_split_selectedFeatures]
        print ('X_patients.shape after applying selected features: ',X_patients.shape)
        
    if useScaler:
        #get scaler and use
        scaler_split_file='/net/mraid08/export/jafar/Microbiome/Analyses/ShaniBAF/\
predictions2/PNP530_majorPhenotypes/%s/%s_split%s_scaler.pkl' %(resultFolder,phen,split)
        phenotype_split_scaler=pickle.load(open(scaler_split_file,"rb"))
        
        X_patients = pd.DataFrame(index=X_patients.index,columns=X_patients.columns, data=scaler.transform(X_patients))    

    #get target DF:
    Y=pd.read_excel(Y_file).set_index('BD')
    Y_patients=Y.loc[Cardio126,:]
#     Y_patients[phen]=Y_patients.fillna(Y_patients[phen].median())
#     print ('Y_patients.shape: ',Y_patients.shape)
    
    #predict phenotype:
    if task=='reg':
        y_pred_patients_split = phenotype_split_model.predict(X_patients)
    else:
        y_pred_patients_split = phenotype_split_model.predict_proba(X_patients)[:,1]
#     print ('first 5 predicted values: ',y_pred_patients_split[:5])
    
        
    #### calculate spearman and pearson for patients and comapre to controls:
    prediction_df_patients_split=pd.DataFrame(data=y_pred_patients_split,index=Cardio126,columns=['y_pred_patients_split%s' %split])
    prediction_df_patients_split=pd.merge(prediction_df_patients_split,Y_patients,how='left',left_index=True,right_index=True)
#     print ('prediction_df_patients_split.shape: ',prediction_df_patients_split.shape)
#     print  prediction_df_patients_split.head(10)

    pred_and_real_df=prediction_df_patients_split; pred_col='y_pred_patients_split%s' %split
#     print 'pred_and_real_df'
#     print pred_and_real_df.head()
#     print ('phen: ',phen)
    r_pear,p_pears, r_spear,p_spears,prevalence,AUC,precision_Recall,kappa,f1score,f1scoreMacro=calc_pred_performance(pred_and_real_df,task,pred_col,phen)
    
    
    return r_pear,p_pears, r_spear,p_spears,prevalence,AUC,precision_Recall,kappa,f1score,f1scoreMacro,prediction_df_patients_split
    
```

```python
def predict_phen_for_patients_basedOnHealthy(phen,resultFolder,X_file,Y_file,task,binary_threshold=0.5,useSelectedFeatures=False,useScaler=False,predName=''):
    
    df=pd.DataFrame()
    pred_values_df=pd.DataFrame()
    for i in  range(1,11):
        print i
        split = i

        r_pear,p_pears, r_spear,p_spears,prevalence,AUC,precision_Recall,kappa,f1score,f1scoreMacro,prediction_df_patients_split=\
        predict_phen_for_patients_basedOnHealthy_1split(phen,resultFolder,X_file,Y_file,split,task,binary_threshold=0.5,useSelectedFeatures=useSelectedFeatures,useScaler=useScaler)
        
        pred_values_df=pd.merge(pred_values_df,pd.DataFrame(prediction_df_patients_split['y_pred_patients_split%s' %i]),how='outer',left_index=True,right_index=True)

        if task == 'reg':  performance =r_pear
        else: performance = AUC

        df.loc[i,'perf_metric'] = performance

    df=df.sort_values(by='perf_metric')
    
    best=df.index[0];  worse=df.index[-1];   median_ind=int(round((len(df)/2)))
    median=df.index[median_ind]
    
    print pred_values_df.head()
    print('best,worse,median: ', best,worse,median)
    
    pred_values_df=pred_values_df[['y_pred_patients_split%s' %best,'y_pred_patients_split%s' %worse,'y_pred_patients_split%s' %median]]
    pred_values_df['average pred']=pred_values_df.mean(axis=1)
    
    
    pred_and_real=pd.merge(pd.DataFrame(pred_values_df['average pred']),pd.DataFrame(prediction_df_patients_split[phen]),how='inner',left_index=True,right_index=True)
#     print 'pred_and_real:'
#     print pred_and_real.head()

#     pred_and_real['average pred'].to_pickle('')
    
    pred_and_real_df=pred_and_real; pred_col='average pred'
    r_pear,p_pears, r_spear,p_spears,prevalence,AUC,precision_Recall,kappa,f1score,f1scoreMacro=calc_pred_performance(pred_and_real_df,task,pred_col,phen)
    s,p=stats.ttest_rel(pred_and_real[phen],pred_and_real['average pred'], axis=0, nan_policy='omit')
    print ('paired-t test: s,p=',s,p)
                        
    pred_and_real['res']=pred_and_real['average pred'] - pred_and_real[phen]
    print 'redisual descriptive stats:'
    print pred_and_real['res'].describe()
    print ''
    print ('number of too high predictions: ',pred_and_real[pred_and_real['res']>0].shape[0])
    print ('number of too low predictions: ',pred_and_real[pred_and_real['res']<0].shape[0])
                        
    
    df2=pd.DataFrame()
    df2.loc[0,'r_pear']=r_pear
    df2.loc[0,'p_pears']=p_pears
    df2.loc[0,'r_spear']=r_spear
    df2.loc[0,'p_spears']=p_spears
    df2.loc[0,'prevalence']=prevalence
    df2.loc[0,'AUC']=AUC
    df2.loc[0,'precision_Recall']=precision_Recall
    df2.loc[0,'kappa']=kappa
    df2.loc[0,'f1score']=f1score
    df2.loc[0,'f1scoreMacro']=f1scoreMacro
    df2.loc[0,'paired t test s']=s
    df2.loc[0,'paired t test p']=p
    df2.loc[0,'# too high preds']=pred_and_real[pred_and_real['res']>0].shape[0]
    df2.loc[0,'# too low preds']=pred_and_real[pred_and_real['res']<0].shape[0]
    
    df2.to_excel(PRED_RESULTS_DIR+'Cardio126_phens_basedOnHealthy/%s_prediction_performance_%s.xlsx' %(phen,predName))
    
    ###add res info to phenotypic info:
    allPhensWithZscore=pd.read_excel('/net/mraid08/export/jafar/Microbiome/Analyses/TCR/phenotypes_byBD/PNP530Cardio126_Age_Gender_Male_HbA1C_SmokingStatus_Yes_BMI_HDL_Smoking_ever_withPredProba.xlsx').set_index('BD').loc[Cardio126,:]
    allPhensWithZscore_withRes=pd.merge(allPhensWithZscore,pd.DataFrame(pred_and_real['res'].rename(phen+'_res')),how='left',left_index=True,right_index=True)
    allPhensWithZscore_withRes=allPhensWithZscore_withRes.sort_values(by=phen+'_res')
    print ('allPhensWithZscore_withRes.head=',allPhensWithZscore_withRes.head(10))
    print ('allPhensWithZscore_withRes.tail=',allPhensWithZscore_withRes.tail(10))
    allPhensWithZscore_withRes.to_excel(PRED_RESULTS_DIR+'Cardio126_phens_basedOnHealthy/%s_pred_residuals_withPhens_%s.xlsx' %(phen,predName))
 
    ### plot correlation:
    
    
    #plot all values:
    fig,ax=plt.subplots()

    ax, nsamples,r,p,text, handles, labels=plot_corr(pred_and_real[phen],pred_and_real['average pred'],data1name='Real '+phen ,data2name='Predicted '+phen,
    ax=ax,title=phen+ ' prediction correlation - Patients',corrType='pearson',toAnnotate=True)

    min_val=int(round(pred_and_real[[phen,'average pred']].min().min()))
    max_val=int(round(pred_and_real[[phen,'average pred']].max().max()))

    ax.plot(range(min_val,max_val),range(min_val,max_val),color='red')
    plt.show()

    fig.savefig(PRED_RESULTS_DIR+'Cardio126_phens_basedOnHealthy/%s_prediction_corr.png' %phen  )

    #plot only values that are in the range of healthy vales
    Y=pd.read_excel(Y_file).set_index('BD')
    Y_healthy=Y.loc[PNP530,:]
    min_healthy_real_val=Y_healthy[phen].min()
    max_healthy_real_val=Y_healthy[phen].max()
    print ('min,max healthy real vals:',min_healthy_real_val,max_healthy_real_val)
    print pred_and_real.shape
    pred_and_real_lim=pred_and_real[(pred_and_real[phen]>=min_healthy_real_val) & (pred_and_real[phen]<=max_healthy_real_val)]
    print pred_and_real_lim.shape
    fig2,ax2=plt.subplots()
   
                    
    ax2, nsamples,r,p,text, handles, labels=plot_corr(pred_and_real_lim[phen],pred_and_real_lim['average pred'],data1name='Real '+phen ,data2name='Predicted '+phen,
    ax=ax2,title=phen+ ' prediction correlation - Patients\nOnly healthy subject value range',corrType='pearson',toAnnotate=True)

    min_val=int(round(pred_and_real_lim[[phen,'average pred']].min().min()))
    max_val=int(round(pred_and_real_lim[[phen,'average pred']].max().max()))

    ax2.plot(range(min_val,max_val),range(min_val,max_val),color='red')
    plt.show()
    fig2.savefig(PRED_RESULTS_DIR+'Cardio126_phens_basedOnHealthy/%s_prediction_corr_limToHealthyVals.png' %phen )
    
    return pred_and_real
    
    
    
```

```python
def compare_healthy_patients_predictions_grouped(phen,real_data_path,pred_data_file,bin_edges,bin_names,ax=None,isBinary=False):
    
    if isBinary:
        bin_edges=[-1,0,1]
    
    ## get predicted and real data for healthy:

    real_data=pd.read_excel(real_data_path).set_index('BD')[phen].dropna()

    pred_data_PNP530=pd.read_pickle(pred_data_file)[phen].dropna()


    #merge for healthy:
    pred_and_real_PNP530=pd.merge(pd.DataFrame(real_data.rename(phen)),pd.DataFrame(pred_data_PNP530.rename('pred '+phen).astype(float)),how='inner',left_index=True,right_index=True)
    print ('pred_and_real_PNP530.shape: ',pred_and_real_PNP530.shape)
    print pred_and_real_PNP530.head()

    ### get predicted and real data for patients:
    pred_and_real_Cardio126=pd.read_pickle(PRED_RESULTS_DIR+'Cardio126_phens_basedOnHealthy/pred_and_real_%s.pkl' %phen)[['average pred',phen]]
    pred_and_real_Cardio126=pred_and_real_Cardio126.rename(columns={'average pred':'pred '+phen})
    print ('pred_and_real_Cardio126: ',pred_and_real_Cardio126.shape)
    pred_and_real_Cardio126.head()
    
    ##plot:
    if ax is None:
        fig,ax=plt.subplots(figsize=(5,5))


    values_dict={}

    for item in [('Healthy',pred_and_real_PNP530,'grey','lightgrey'),('Patients',pred_and_real_Cardio126,'darkred','red')]:
        df=item[1]
        df_name=item[0]
        color_real=item[2]
        color_pred=item[3]

    #     print df.columns.tolist()
        if isBinary:
            fig_range=range(2)
        else:
            fig_range=range(1,len(bin_edges))

        df['Binned '+phen]=pd.cut(df[phen],bin_edges,labels=fig_range)
        grouped=df.groupby('Binned '+phen).agg([np.mean,np.std,np.size])
        grouped.columns = [' '.join(col).strip() for col in grouped.columns.values]

        for n in fig_range:
            values_dict[df_name+str(n)]=df[(df[phen]>=bin_edges[n-1]) & (df[phen]<bin_edges[n])]['pred '+phen].tolist()

        print df_name
        print grouped
        
        if not isBinary:
            ax.errorbar(fig_range,grouped[phen+ ' mean'],yerr=grouped[phen+ ' std'], fmt='^',color=color_real,alpha=1,linewidth=2,markersize=4,
                    label=df_name+' (real)')
        ax.errorbar(fig_range,grouped['pred '+phen+' mean'],yerr=grouped['pred '+phen+' std'], fmt='o',color=color_pred,alpha=0.5,linewidth=4,markersize=8,
                   label=df_name+' (pred.)')
        ax.set_xticks(fig_range)
        ax.set_xticklabels(bin_names,rotation=45,fontsize='medium')
    min_values=[]; max_values=[]
    for n in fig_range:
        data1=values_dict['Healthy'+str(n)]
        data2=values_dict['Patients'+str(n)]
#         min_values.append(np.min(data1+data2))
#         max_values.append(np.max(data1+data2))
        MW_s, MW_p=mannwhitneyu(data1,data2)
        print ('MW test: %s bin ' %phen,bin_names[n-1],MW_s, MW_p)

        t_s, t_p=ttest_ind(data1,data2)
        print ('t test: %s bin ' %phen,bin_names[n-1],t_s, t_p)
        print ''

        ax.legend(loc='upper left')
        ax.set_ylabel(phen+ '',fontsize='large')
        ax.set_xlabel(phen+ ' Group',fontsize='large')

#     ymin=adjusted_rounddown(np.min(min_values))
#     ymax=adjusted_roundup(np.min(max_values))
      
        
    xmin=fig_range[0]-0.5; xmax=fig_range[-1]+0.5
    
    ax.set_xlim(xmin,xmax)
#     ax.set_ylim(ymin,ymax)

    return ax
    
    
```

```python
def gen_shap_summary_to_axes(ax, fig, target_name, nTopFeatures, shap_df,
                              features_df, jitter=None, sample_num=None,
                               feature_name_list=None, scalingMethod='minmax',
                               addColorBar=True, alpha=1):
    # ##optional values for scalingMethod: 'minmax','scale','perc',None
    
    
     # calcualte feature order by sum of abs shap values and take top n values:
    if nTopFeatures is not None:
        topN_features = shap_df.abs().sum().sort_values(ascending=False)[:nTopFeatures].index.tolist()
        print topN_features
        topN_features_new = edit_feature_names(topN_features)
        print topN_features_new
    else:
        topN_features = shap_df.abs().sum().sort_values(ascending=False)
    
    # if not None, select limited number of samples to present (used to edit the function and test it quickly)
    if sample_num is None:
        sample_num = shap_df.shape[0]
       
    features_df_topN = features_df.loc[:, topN_features].iloc[:sample_num, :]
    shap_df_topN = shap_df.loc[:, topN_features].iloc[:sample_num, :]
    print ('sample_num= ', sample_num)

    # scale feature data- to enable color coding to show samples high and low in each feature:
    if scalingMethod is not None:
        if scalingMethod == 'minmax':
            # using min-max scaler 
            scaler = preprocessing.MinMaxScaler(copy=True, feature_range=(0, 1))
            features_df_topN_scaled = pd.DataFrame(index=features_df_topN.index, columns=features_df_topN.columns, data=scaler.fit_transform(features_df_topN))
        if scalingMethod == 'scale':
            # center to the mean and to get unit variance
            scaler = preprocessing.scale(copy=True)
            features_df_topN_scaled = pd.DataFrame(index=features_df_topN.index, columns=features_df_topN.columns, data=scaler.fit_transform(features_df_topN))
        elif scalingMethod == 'perc':
            features_df_topN_scaled = features_df_topN / features_df_topN.sum()
        else:
            print 'scaling method was not identified, no scaling was done'
            features_df_topN_scaled = features_df_topN
    else:
        print 'no scaling was done'
        features_df_topN_scaled = features_df_topN
    
    # transform to longform tables and merge:
    features_df_topN_scaled_long = pd.melt(features_df_topN_scaled.reset_index(), id_vars='index', value_name='feature_value')
    shap_df_topN_long = pd.melt(shap_df_topN.reset_index(), id_vars='BD', value_name='shap_value')
    merged = pd.merge(features_df_topN_scaled_long, shap_df_topN_long, how='inner', left_on=['index', 'variable'], right_on=['BD', 'variable']).rename(columns={'index':'Sample', 'variable':'feature'})
    
    print ('features_df_topN.shape=', features_df_topN.shape)
    print ('features_df_topN_scaled.shape=', features_df_topN_scaled.shape)
    print ('features_df_topN_scaled_long.shape=', features_df_topN_scaled_long.shape)
    print ('shap_df_topN.shape=', shap_df_topN.shape)
    print ('shap_df_topN.shape=', shap_df_topN.shape)
    print ('merged.shape=', merged.shape)
    print ('count shap values per feature:')
    print merged.head(20)
    print 'merged groupby feature - count:'
    print merged[['feature', 'shap_value']].groupby('feature').count()
    



    # ##plot:
    ax.grid(False)
    # generate unuseful scatter plot just to give colorbar to color range...
    stam = plt.figure()
    stam1 = plt.scatter(merged['shap_value'], merged['feature_value'] , c=merged['feature_value'], cmap='RdBu_r')
    plt.clf()
    
    # generate specific ax for the colorbar and plot the colorbar:
    if addColorBar:
        divider = make_axes_locatable(ax)
        ax_cb = divider.new_horizontal(size="4%", pad=0.05)
        fig.add_axes(ax_cb)
        cb1 = fig.colorbar(stam1, cax=ax_cb, orientation='vertical')
        ax_cb.yaxis.set_ticks([0, 0.5, 1])
        ax_cb.yaxis.set_ticklabels([0, 0.5, 1],fontsize=18)
        ax_cb.set_ylabel('Norm Feature Value')
        
    # plot the scatter plot itself:
    plot = sns.stripplot(x='feature', y='shap_value', hue='feature_value', data=merged,
                      palette='RdBu_r',
                     alpha=alpha, ax=ax, s=5, jitter=jitter)
        
#         .swarmplot   edgecolor='grey'  ,linewidth=2
#     else:
#         plot = sns.stripplot(x='feature', y='shap_value', hue='feature_value', data=merged,
#                       palette='RdBu_r',
#                        edgecolor='none', alpha=.4, ax=ax, s=5)
    
    plot.get_legend().set_visible(False)
    ax.set_xlabel("")
    ax.set_ylabel('Shap Value (%s)' % target_name)
    ax.set_xticklabels(topN_features_new)  # change ticklables to readable feature names
    plt.setp(ax.get_xticklabels(), rotation=90, ha="left",fontsize=18, fontweight='bold')
    
    # color ticklables according to their content:
    for n, lab in enumerate(ax.get_xticklabels()):
        str_lab = str(lab)
#         print n,str(lab)
        if ('insertion' in str_lab) or ('deletion' in str_lab) or ('length' in str_lab):
            plt.setp(ax.get_xticklabels()[n], color='green')
        elif ('Norm' in str_lab) or ('Top' in str_lab) or ('count' in str_lab)\
or ('Shannon' in str_lab) or ('Berger' in str_lab) or ('Simpson' in str_lab) :
            plt.setp(ax.get_xticklabels()[n], color='red')
        elif ('seq #' in str_lab) or ('seq freq.' in str_lab) or ('annotate' in str_lab) :
            plt.setp(ax.get_xticklabels()[n], color='orange')
        elif ('non-prod' in str_lab) :
            plt.setp(ax.get_xticklabels()[n], fontstyle='italic')

#             plt.setp(ax.get_xticklabels()[n], r'\underline{Parameters}: ')
            
    ax.set_xticks([x + 0.25 for x in ax.get_xticks()])
    ax.tick_params(axis='x', direction="in", pad=-3, labelright=True,size=18)


    return ax, fig
```

# associations between age, gender and HbA1C and top 100 sequences:

```python
def analyze_tcr_phen_relations(phenDF,TCRdf_binary,nShared_perphen,nShared_perTCR,binary_zero=0.0001,phen_name='MbSpecies'):
    
    #this function is a generalized form of the function analyze_tcr_phen_relations
    # in this function, the phen can be microbiome relative abundance table or any other phenotypic table
    # the function will calculate the fisher and mann-whitney associations between each TCR sequence/cluster in the TCRdf table, which is binary,
    # and the phenDF table which is not binary
    # for fisher test, a binary cut off for the phen should be selected. for continuous phenotypes, consdier using the mean or median. for microbiome 
    #RA data, this is usually 0.0001
    ########
    
    #filter phen and generate binary df:
    sample_count=phenDF.count()
    phen_filtered=phenDF.loc[:,sample_count[(sample_count>=nShared_perphen)&(sample_count<len(phenDF)+1)].index].fillna(0)
    print ('phenDF shape after filtering only for phen that appear in more than %s: ' %nShared_perphen,phen_filtered.shape)
#     print phen_filtered.iloc[:4,:4]
    phen_filtered_binary=(phen_filtered>binary_zero).astype(int)
    print 'binary:'
#     print phen_filtered_binary.iloc[:4,:4]
    
    #filter TCRs:
    cols_to_use=[col for col in TCRdf_binary if (TCRdf_binary[col].sum()>=nShared_perTCR) & (TCRdf_binary[col].sum()<len(TCRdf_binary))]
    TCRdf_binary_filtered=TCRdf_binary[cols_to_use]
    print ('TCRdf shape after filtering only for seqs that appear in more than %s: ' %nShared_perTCR,TCRdf_binary_filtered.shape)
#     print TCRdf_binary_filtered.iloc[:4,:4]

    print ('total number of tests: ',TCRdf_binary_filtered.shape[1]*phen_filtered.shape[1] )
    
    #run over all seqs and phen and calculate MW and fisher test:
    summary_df=pd.DataFrame()
    count=0
    for i, seq in enumerate(TCRdf_binary_filtered.columns):
        print i, seq
    #     print i,seq
        for j, s in enumerate(phen_filtered.columns):
            
            #calculate mann-whitney:
            merged1=pd.merge(pd.DataFrame(TCRdf_binary_filtered[seq]),pd.DataFrame(phen_filtered[s]),
                                            how='inner',left_index=True,right_index=True)
#             print ('merged1 shape: ', merged1.shape)
#             print 'merged1'
#             print merged1.head()
            data={}
            names=[]
            for name,group in merged1.groupby(seq):
                data[name]=group[s].tolist()
                names.append(name)
            try:
                MW_s, MW_p=mannwhitneyu(data[names[0]],data[names[1]])          
                mean_phenFreq_seqAbs=np.mean(data[0])
                mean_phenFreq_seqPres=np.mean(data[1])
            except:
                print 'MW test failed, probably this clusters doesnt exist in samples'
                MW_s=np.nan; MW_p=np.nan; mean_phenFreq_seqAbs=np.nan; mean_phenFreq_seqPres=np.nan
            
            #calculate Fisher:
            merged2=pd.merge(pd.DataFrame(TCRdf_binary_filtered[seq]),pd.DataFrame(phen_filtered_binary[s]),
                                            how='inner',left_index=True,right_index=True)
#             print 'merged2'
#             print merged2.head()
            if (merged2[s].sum()<len(merged2)) | (merged2[seq].sum()<len(merged2)):
                tab = pd.crosstab(merged2[seq],merged2[s], dropna=False).fillna(0)
                seqAbs_phenAbs=tab.iloc[0,0]
                seqAbs_phenPres=tab.iloc[0,1]
                seqPres_phenAbs=tab.iloc[1,0]
                seqPres_phenPres=tab.iloc[1,1]
                
    #             print (tab)
                try:
                    fisher_OR, fisher_p = fisher_exact(tab, alternative='two-sided')
                except:
                    print ('couldnt execute fisher test for seq %s and phen %s' %(seq,s))
#                     print (tab)
                    fisher_OR= 999999; fisher_p=1
            else:
                fisher_OR= 999999; fisher_p=1
                seqAbs_phenAbs=np.nan; seqAbs_phenPres=np.nan
                seqPres_phenAbs=np.nan; seqPres_phenPres=np.nan
              
            #update summary_df:
            summary_df.loc[count,'seq']=seq
            summary_df.loc[count,phen_name]=s
            summary_df.loc[count,'mean_phenFreq_seqAbs']=mean_phenFreq_seqAbs
            summary_df.loc[count,'mean_phenFreq_seqPres']=mean_phenFreq_seqPres
            summary_df.loc[count,'MW_s']=MW_s
            summary_df.loc[count,'MW_p']=MW_p
            summary_df.loc[count,'MW_s']=MW_s
            summary_df.loc[count,'MW_p']=MW_p
            summary_df.loc[count,'seqAbs_phenAbs']=seqAbs_phenAbs
            summary_df.loc[count,'seqAbs_phenPres']=seqAbs_phenPres
            summary_df.loc[count,'seqPres_phenAbs']=seqPres_phenAbs
            summary_df.loc[count,'seqPres_phenPres']=seqPres_phenPres
            summary_df.loc[count,'fisher_OR']=fisher_OR
            summary_df.loc[count,'fisher_p']=fisher_p

            #update count:
            count=count+1

    summary_df=add_corrected_pValues(summary_df,pValueColumn='MW_p',nTests=summary_df.shape[0],FDR=0.1)
    
    summary_df=add_corrected_pValues(summary_df,pValueColumn='fisher_p',nTests=summary_df.shape[0],FDR=0.1)
    summary_df=summary_df.sort_values(by='MW_p')


    return summary_df
```
