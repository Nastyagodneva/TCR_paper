# generate shap-values heatmap

```python
#phen_list=list of tuples. the first item in the tuple is the phenotype name (with spaces!) and the second one is the 
# the folder name

def extract_featImpData_plotHeatMap(phen_list,nTop,pred_dir,output_dir,transpose=False):
#     pred_dir='/net/mraid08/export/genie/Lab/Personal/ShaniBAF/TCR_real_data/predictions2/Cardio126_diseasePhenotypes/'

    shap_summary_df=pd.DataFrame()

    for phen in phen_list:
        print phen[0]
        f1=pred_dir+phen[1]+'/coef_values.pkl'
        if ('lin' in phen[1]) or ('log' in phen[1]):
            f1=pred_dir+phen[1]+'/coef_values.pkl'
        else:
            f1=pred_dir+phen[1]+'/shap_values.pkl'

        impFeats=pd.read_pickle(f1)[phen[0]]
        impFeats2=impFeats.abs().sum()
        impFeats2Normed=(impFeats2/impFeats2.max()).rename(phen[0]+'_normed')
        impFeats2sign=(impFeats.sum().apply(lambda x: -1 if x<0 else 1)).rename(phen[0]+'_sign')
    #     print impFeats2sign.min()
        impFeats2Total=pd.merge(pd.DataFrame(impFeats2Normed),pd.DataFrame(impFeats2sign),how='inner',left_index=True,right_index=True)
        impFeats2Total[phen[0]]=impFeats2Total[phen[0]+'_normed']*impFeats2Total[phen[0]+'_sign']
        impFeats2Total=impFeats2Total.sort_values(by=(phen[0]+'_normed'),ascending=False)
    #     print impFeats2Total.head(20)

        shap_summary_df=pd.merge(shap_summary_df,pd.DataFrame(impFeats2Total[phen[0]][:nTop]),
                                 how='outer',left_index=True,right_index=True)
    shap_summary_df.to_excel(output_dir+'shap_summary_df_%s.xlsx' %nTop)

    #plot heatmap:
    import seaborn
    if transpose: 
        shap_summary_df=shap_summary_df.T
        figsize=(50,28)
    else:
        figsize=(18,44)
    fig,ax=plt.subplots(figsize=figsize)
    g=seaborn.heatmap(shap_summary_df.fillna(0).round(3).drop_duplicates(keep=False),ax=ax,cbar_kws={"shrink": .82})
    cbar = ax.collections[0].colorbar
    cbar.ax.tick_params(labelsize='x-large')
    plt.setp(g.get_yticklabels(), rotation=0,fontsize='xx-large')
    plt.setp(g.get_xticklabels(), rotation=90,fontsize='large')
    plt.show()

    figfile=output_dir+'shap_heatmap_%s' %nTop
    fig.savefig(figfile,dpi=300)

    return shap_summary_df
```

```python
def plot_hist_for_2000perms(phen,dir1,metric,ax):
    sum_df=pd.read_excel('/net/mraid08/export/jafar/Microbiome/Analyses/ShaniBAF/predictions2/Cardio126_diseasePhenotypes/FinalCandidates/%s/sum_df_%s.xlsx' %(dir1,phen))
    sum_df_shuf=sum_df[(sum_df.index.str.contains('shuf'))&(~sum_df.index.str.contains('mean'))]
    print sum_df_shuf.head()
    
    if metric is None:
        metric=sum_df.columns.tolist()[0]
        print ('metric: ',metric)
    real=sum_df.loc['real',metric]
    predicted=sum_df.loc['predAgeGender',metric]
    
    if (metric=='f1scoreMacro') | (metric=='AUC'):
        xmin=0
        xmax=1
        n=11
    else:
        xmin=-1
        xmax=1
        n=21
    xticks=np.linspace(xmin, xmax, 11)
    bins=np.linspace(xmin, xmax, 21)
    
    sum_df_shuf[metric].hist(ax=ax,color='grey', alpha=0.5,bins=bins)
    ax.set_xticks(xticks)
    
    ymin,ymax=ax.get_ylim()
    arrow_top=ymin+((ymax-ymin)/2)
    
    ax.annotate("real", xy=(real, 0), xytext=(real, arrow_top),arrowprops={'facecolor':'red'},fontweight='bold')
    ax.annotate("predicted", xy=(predicted, 0), xytext=(predicted, arrow_top),arrowprops={'facecolor':'blue'},fontweight='bold')
    
    
    return fig,ax,metric
```

```python
phen='DischargeClopidogrel'
dir1='DischargeClopidogrel_logReg20'
metric='f1scoreMacro'

fig,ax=plt.subplots()
fig,ax,metric=plot_hist_for_2000perms(phen,dir1,metric,ax)

plt.show()
```

```python
def get__limited_sharing(min_share,max_share):
    
    print ('initial shape is: ',newX_onlySeqs_Cardio126.shape)
    
    min_share_number=math.ceil(len(newX_onlySeqs_Cardio126) * min_share)
    max_share_number=math.ceil(len(newX_onlySeqs_Cardio126) * max_share)
    
    print ('min_share_number: ',min_share_number)
    print ('max_share_number: ',max_share_number)
    
    newX_onlySeqs_Cardio126_limited=newX_onlySeqs_Cardio126.loc[:, newX_onlySeqs_Cardio126.count()>=min_share_number]
    newX_onlySeqs_Cardio126_limited=newX_onlySeqs_Cardio126_limited.loc[:, newX_onlySeqs_Cardio126.count()<max_share_number]
    
    print ('final shape is: ',newX_onlySeqs_Cardio126_limited.shape)
    newX_onlySeqs_Cardio126_limited=newX_onlySeqs_Cardio126_limited.fillna(0)
    
    return newX_onlySeqs_Cardio126_limited
```
