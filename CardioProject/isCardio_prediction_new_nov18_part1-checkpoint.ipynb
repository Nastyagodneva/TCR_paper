### define function:

```python
def predict_phen_for_patients_basedOnHealthy(split,resultFolder,phenotype,task,binary_threshold=0.5):
    
    
    r_pear,y_pears, r_spear,y_spears,prevalence, AUC,precision_Recall,kappa,f1score,f1scoreMacro = np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan
    ### prediction:
    
    #get saved model:
    phenotype_split_model_file='/net/mraid08/export/jafar/Microbiome/Analyses/ShaniBAF/\
predictions2/PNP530_majorPhenotypes/%s/%s_split%s_model.pkl' %(resultFolder,phenotype,split)
    phenotype_split_model=pickle.load(open(phenotype_split_model_file,"rb"))

    #get feature matrix for patients only:
    X_file='/net/mraid08/export/genie/Lab/Personal/ShaniBAF/TCR_real_data/Predictions/featureDFs/\
TCRfeaturesNoT_PCA10percShared_withRels_VDJ_noCorr0-999_nanFilled_noConsts.dat'
    X=pd.read_pickle(X_file)

    with open('/net/mraid08/export/genie/Lab/Personal/ShaniBAF/Sample files/BD lists/Cardio126','rb') as fp:
        Cardio126=pickle.load(fp)
    X_patients=X.loc[Cardio126,:]
    print ('X_patients.shape: ',X_patients.shape)
    
    #get target DF:
    Y_file='/net/mraid08/export/jafar/Microbiome/Analyses/ShaniBAF/predictions2/TargetDFs/PNP530Cardio126_%s.xlsx' %phenotype
    Y=pd.read_excel(Y_file).set_index('BD')
    Y_patients=Y.loc[Cardio126,:]
    Y_patients[phenotype]=Y_patients.fillna(Y_patients[phenotype].median())
    print ('Y_patients.shape: ',Y_patients.shape)
    
    #predict phenotype:
    if task=='reg':
        y_pred_patients_split = phenotype_split_model.predict(X_patients)
    else:
        y_pred_patients_split = phenotype_split_model.predict_proba(X_patients)[:,1]
    print ('first 5 predicted values: ',y_pred_patients_split[:5])
    
        
    #### calculate spearman and pearson for patients and comapre to controls:
    prediction_df_patients_split=pd.DataFrame(data=y_pred_patients_split,index=Cardio126,columns=['y_pred_patients_split%s' %split])
    prediction_df_patients_split=pd.merge(prediction_df_patients_split,Y_patients,how='left',left_index=True,right_index=True)
    print ('prediction_df_patients_split.shape: ',prediction_df_patients_split.shape)
    print  prediction_df_patients_split.head(10)
    
    if task=='reg':


        #get spearman and pearson correlations:
        r_pear,y_pears=pearsonr(prediction_df_patients_split['y_pred_patients_split%s' %split],prediction_df_patients_split[phenotype])
        print ('r_pear,y_pears: ',r_pear,y_pears)
        r_spear,y_spears=spearmanr(prediction_df_patients_split['y_pred_patients_split%s' %split],prediction_df_patients_split[phenotype])
        print ('r_spear,y_spears: ',r_spear,y_spears)
        
    else:
         #### calculate AUC and PR:
        prevalence = float(Y_patients.sum()) / Y_patients.shape[0]
        # AUC
        fpr, tpr, thresholds = metrics.roc_curve(Y_patients, y_pred_patients_split, pos_label=1)
        AUC = metrics.auc(fpr, tpr)
        precision, recall, _ = precision_recall_curve(Y_patients, y_pred_patients_split)
        precision_Recall = metrics.auc(recall, precision)
        
        print ('Binary probability threshold=%s' %binary_threshold)
        y_pred_patients_split_binary=[0 if x<binary_threshold else 1 for x in list(y_pred_patients_split)]    
        # Cohen's Kappa:
        kappa=metrics.cohen_kappa_score(Y_patients,y_pred_patients_split_binary)
        #f1 scores:
        f1score=metrics.f1_score(Y_patients,y_pred_patients_split_binary)
        f1scoreMacro=metrics.f1_score(Y_patients,y_pred_patients_split_binary,average='macro')
        
        print ('prevalence: ',prevalence)
        print ('ROC_auc: ',AUC)
        print ('precision_Recall: ',precision_Recall)
        print ("cohen's kappa: ",kappa)
        print ('f1score:', f1score)
        print ('f1scoremacro:' ,f1scoreMacro)
            
            
        
    
    return r_pear,y_pears, r_spear,y_spears,prevalence,AUC,precision_Recall,kappa,f1score,f1scoreMacro,prediction_df_patients_split
    
```

## check gender prediction in unbalanced healthy population!

```python
def check_gender_prediction_in_unbalanced_healthy_cohort(binary_threshold):
    ### learn model on half of the samples (balanced!)
    #### split cohort:
    from sklearn.model_selection import train_test_split

    X_file='/net/mraid08/export/genie/Lab/Personal/ShaniBAF/TCR_real_data/Predictions/featureDFs/TCRfeaturesNoT_PCA10percShared_withRels_VDJ_noCorr0-999_nanFilled_noConsts.dat'
    X=pd.read_pickle(X_file)
    X_PNP530=X.loc[PNP530,:]
#     print X_PNP530.shape
#     print X_PNP530.iloc[:4,:4]
    Y_file='/net/mraid08/export/genie/Lab/Personal/ShaniBAF/TCR_real_data/predictions2/TargetDFs/PNP530_Gender.xlsx'
    Y=pd.read_excel(Y_file).set_index('BD').loc[PNP530,:]
#     print Y.shape
#     print Y.iloc[:4,:4]
    X_train, X_test, y_train, y_test = train_test_split(X_PNP530, Y, test_size=0.5,stratify=Y)
    
    #### train model:
    model_params={"reg_alpha": 1, "learning_rate": 0.05, "n_estimators": 1200, "subsample": 0.5, "max_depth": 5, "gamma": 0}
    modelObject = XGBClassifier(**model_params)
    modelObject.fit(X_train, y_train)
    
    ### generate non-balanced cohort from the second half and predict gender using the existing model
    y_test_female=y_test[y_test['Gender_Male']==0]
    y_test_male=y_test[y_test['Gender_Male']==1]
    print ('male/female ration in the balanced cohort is: ', float(len(y_test_male))/len(y_test_female))
    y_test_female_sampled=y_test_female.sample(n=11)
    print ('male/female ration in the balanced cohort is: ', float(len(y_test_male))/len(y_test_female_sampled))
    imbalanced_Y=pd.concat([y_test_male,y_test_female_sampled])
#     print imbalanced_Y.shape
    imbalanced_X=X_test.loc[imbalanced_Y.index,:]
#     print imbalanced_X.shape

    ### predict on the new unbalanced healthy cohort:
    y_pred_imbalanced_gender = modelObject.predict_proba(imbalanced_X)[:,1]
    
    #evaluate prediction:
    prevalence = float(imbalanced_Y.sum()) / imbalanced_Y.shape[0]
    print ('prevalence:',prevalence)
    fpr, tpr, thresholds = metrics.roc_curve(imbalanced_Y, y_pred_imbalanced_gender, pos_label=1)
    AUC = metrics.auc(fpr, tpr)
    print ('AUC: ',AUC)
    precision, recall, _ = precision_recall_curve(imbalanced_Y, y_pred_imbalanced_gender)
    Precision_Recall = metrics.auc(recall, precision)
    print ('Precision_Recall: ',Precision_Recall)
    PR_gain_fold=Precision_Recall/prevalence
    print ('PR_gain_fold: ',PR_gain_fold)
    print ('Binary probability threshold=%s' %binary_threshold)
    y_pred_imbalanced_gender_binary=[0 if x<binary_threshold else 1 for x in list(y_pred_imbalanced_gender)]    
    # Cohen's Kappa:
    kappa=metrics.cohen_kappa_score(imbalanced_Y,y_pred_imbalanced_gender_binary)
    print ('kappa=%s' %kappa)
    #f1 scores:
    f1score=metrics.f1_score(imbalanced_Y,y_pred_imbalanced_gender_binary)
    print ('f1score=%s' %f1score)
    f1scoreMacro=metrics.f1_score(imbalanced_Y,y_pred_imbalanced_gender_binary,average='macro')
    print ('f1scoreMacro=%s' %f1scoreMacro)
    
    return AUC,Precision_Recall,PR_gain_fold,kappa,f1score,f1scoreMacro
    
```

```python
AUC_list=[]
PR_gain_fold_list=[]
kappa_list=[]
f1scoreMacro_list=[]

binary_threshold=0.2


for repeat in range(100):
    print repeat
    AUC,Precision_Recall,PR_gain_fold,kappa,f1score,f1scoreMacro=check_gender_prediction_in_unbalanced_healthy_cohort(binary_threshold)
    AUC_list.append(AUC)
    PR_gain_fold_list.append(PR_gain_fold)
    kappa_list.append(kappa)
    f1scoreMacro_list.append(f1scoreMacro)
```

```python
def plot_PR_curve(ax,y_true,y_pred_proba,pos_label,dataset_name='Precision-Recall curve',color='b'):

    from sklearn import metrics
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])
    for _, spine in ax.spines.items():
        spine.set_visible(True)
        spine.set_linewidth(1.)
        spine.set_color('black')

    precision, recall, _ = metrics.precision_recall_curve(y_true.astype(float), y_pred_proba.loc[y_true.index, :].iloc[:,0])
    pr_auc = metrics.auc(recall, precision)
    prevalence=round((y_true.sum() / y_true.shape[0]),2)
    ax.step(recall, precision, color=color, where='post')
    ax.fill_between(recall, precision, step='post', alpha=0.4,
                     color=color, label='%s - AUC = {0:0.3f}'.format(pr_auc) %dataset_name)
    ax.plot([0, 1], [y_true.sum() / y_true.shape[0], y_true.sum() / y_true.shape[0]], color='navy', lw=1, linestyle='--')
    ax.set_xlabel('Recall', fontsize='large', labelpad=-15)
    ax.set_ylabel('Precision', fontsize='large', labelpad=-12)
    ax.set_ylim([0.0, 1.02])
    ax.set_xlim([0.0, 1.0])
#     add_text_at_corner(ax, 'PR=%0.2g' % pr_auc, 'bottom left', fontsize='large', weight='bold')
#     if n==0:
#         ax.text(0.02,(y_true.sum() / y_true.shape[0])+0.02,'prevalence=%s' %prevalence,fontsize='large')
#         x, y, s, fontdict=None, withdash=False, **kwargs
    
    return ax,pr_auc
            
```

### Find important features by comparing isCardio predictions shap values to age-prediction shap values and calculating residuals

```python
#X and Y shap values should be dataframes with one column each that contains the shap values to be presented on the X axis (the ones we want to 'substract')
# and those to be presented on the Y axis (the ones we want to substract from)
#### this function currently fits 'isCardio' prediction only


def calc_and_plot_shap_res(X_shap_value,Y_shap_values,Xname,Yname,figFolder,nToShow=10):
    assert len(X_shap_value) == len(Y_shap_values) 
    X_shap_value=X_shap_value.loc[Y_shap_values.index]

    #build a linear regression model to predict shap values for isCarido prediction from the shap values for age prediction :
    lm = LinearRegression()  # intercept=True, normalize=False
    lm.fit(X_shap_value,Y_shap_values)

    y_pred = pd.DataFrame(index=X_shap_value.index.tolist(), data=lm.predict(X_shap_value),columns=[Yname+'_pred'])

    #calculate residuals and sort features by residual size:
    merged = pd.merge(Y_shap_values, y_pred, how='inner', left_index=True, right_index=True)
    merged=pd.merge(merged,X_shap_value,how='left',left_index=True,right_index=True)
    merged['res'] = merged[Yname] - merged[Yname+'_pred']
    merged=merged.sort_values(by='res',ascending=False)
    colorList=['r' if n<nToShow else 'black' if n<(len(merged)-2*nToShow) else 'b' for n in range(len(merged))]

    #plot correlation between shap values for age prediction and shap values for isCardio prediction
    #mark features with the highest residuals

    data1=merged[Xname].tolist()
    data2=merged[Yname].tolist()
    data1name=Xname
    data2name=Yname
    
    fig=plt.figure(figsize=(12,5))
    fig.suptitle('Top contributing features for disease classification',fontsize='xx-large')
    ax1=fig.add_subplot(1,2,1)

    ax1, nsamples,r,p,text=plot_corr(data1,data2,data1name,data2name,ax1,title=None,corrType='pearson',
                                     toAnnotate=True,scatter_kws={'alpha':0.5,'s':60,'color':colorList})
    xlim=int(math.ceil(np.max(data1) / 100.0)) * 100 #find the next hundred
    ylim=int(math.ceil(np.max(data2) / 100.0)) * 100 #find the next hundred
    ax1.set_xlim(0,xlim)
    ax1.set_ylim(0,ylim)
    
    ax2=fig.add_subplot(1,3,3)
    #plot 10 most contributing features:
    merged['res'][:nToShow].plot(kind='barh',color='red',ax=ax2)
    ax2.set_xlabel('Shap residuals',fontsize='x-large')
#     fig.subplots_adjust(wspace=0.4)

    figFile=figFolder+Yname+'_normedTo_'+Xname+'.png'
    fig.savefig(figFile,dpi=300)
    plt.show()


```
